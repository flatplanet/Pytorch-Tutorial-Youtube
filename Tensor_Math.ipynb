**TENSOR MATH OPERATIONS**


*   Add, Subtract, Multiply, Divide, Remainders, Exponents
*   Shorthand and uh, Longhand
*   Reassignment 



```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
# Addition
tensor_a + tensor_b
```

```python
# Addition Longhand 
torch.add(tensor_a, tensor_b)
```

```python
# Subtraction 
tensor_b - tensor_a
```

```python
# Sub Function
torch.sub(tensor_b, tensor_a)
```

```python
# Multiplication
tensor_a * tensor_b
```

```python
# Mul longhand
torch.mul(tensor_a, tensor_b)
```

```python
# Division
tensor_b / tensor_a
```

```python
# Div Longhand
torch.div(tensor_b, tensor_a)
```

```python
# Remainder Modulus
tensor_b % tensor_a
```

```python
# Remainder longhand
torch.remainder(tensor_b, tensor_a)
```

```python
# Exponents / power
torch.pow(tensor_a, tensor_b)
```

```python
# Another way to write longhand
tensor_a.add(tensor_b)
```

```python
# Reassignment_
tensor_a + tensor_b
```
